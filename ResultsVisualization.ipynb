## Load saved model result variables
Use of pickle

```python
import pickle
from tensorflow.keras.callbacks import History
```

```python
# Load the CNN model results
with open('/content/drive/MyDrive/DATA_4380/history_simpleCNN.pkl', 'rb') as f:
    history_simpleCNN = pickle.load(f)

# Convert variable from dict to History object with using pickle
# Create a new instance of the History class
new_history_simpleCNN = History()

# Assign the dictionary values to the history attribute
new_history_simpleCNN.history = history_simpleCNN
```

```python
# Load the VGG16 model results
with open('/content/drive/MyDrive/DATA_4380/history_vgg16.pkl', 'rb') as f:
    history_vgg16 = pickle.load(f)

# Convert variable from dict to History object with using pickle
# Create a new instance of the History class
new_history_vgg16 = History()

# Assign the dictionary values to the history attribute
new_history_vgg16.history = history_vgg16
```

```python
# Load the ResNET model results
with open('/content/drive/MyDrive/DATA_4380/history_resnet.pkl', 'rb') as f:
    history_resnet = pickle.load(f)

# Convert variable from dict to History object with using pickle
# Create a new instance of the History class
new_history_resnet = History()

# Assign the dictionary values to the history attribute
new_history_resnet.history = history_resnet
```

## Analyze and Visualize Training Results
plot_training_history() function

```python
import matplotlib.pyplot as plt

def plot_training_history(history):
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))

    # Plot training & validation accuracy values
    ax[0].plot(history.history['accuracy'])
    ax[0].plot(history.history['val_accuracy'])
    ax[0].set_title('Model accuracy')
    ax[0].set_ylabel('Accuracy')
    ax[0].set_xlabel('Epoch')
    ax[0].legend(['Train', 'Validation'], loc='upper left')

    # Plot training & validation loss values
    ax[1].plot(history.history['loss'])
    ax[1].plot(history.history['val_loss'])
    ax[1].set_title('Model loss')
    ax[1].set_ylabel('Loss')
    ax[1].set_xlabel('Epoch')
    ax[1].legend(['Train', 'Validation'], loc='upper left')

    plt.show()
```

## Plot results from all three models
Basic CNN, VGG16, RESNET

```python
# Plot Results from CNN Model using visualization function
plot_training_history(new_history_simpleCNN)
```

```python
# Plot Results from VGG16 Model using visualization function
plot_training_history(new_history_vgg16)
```

```python
# Plot Results from ResNET Model using visualization function
plot_training_history(new_history_resnet)
```
